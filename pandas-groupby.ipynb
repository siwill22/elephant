```python
# how to use groupby

import pandas as pd
import numpy as np
```

```python
random_numbers = np.random.randn(1000)

df = pd.DataFrame(np.vstack((np.round(random_numbers*10),random_numbers)).T, columns=['Nround', 'N'])

df
```

```python
grouped_points = df.groupby(['Nround']) #.size().reset_index().rename(columns={0:'count'})

for group in grouped_points.groups:
    group_values = grouped_points.get_group(group)
    print(group, len(group_values), group_values.N.min(), group_values.N.max())
    
```

```python
df.groupby('N').mean()
```
